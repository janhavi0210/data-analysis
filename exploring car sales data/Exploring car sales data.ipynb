```python
#cleaning column names
def changeColumn(column):
    if column=='yearOfRegistration':
        column='registration_year'
        return column
    elif column=='monthOfRegistration':
        column='registration_month'
        return column
    elif column=='notRepairedDamage':
        column='unrepaired_damage'
        return column   
    elif column=='dateCreated':
        column='ad_created'
        return column
          
    else:
        column_separated=[]
        
        for i in range(len(column)):
            if column[i].isupper()==False:
                column_separated.append(column[i])
            else:
                column_separated.append('_')
                column_separated.append(column[i].lower())
        column=''.join(column_separated)        
        return column
                
changed_dictionary={}
for col in autos.columns:
    new_col=changeColumn(col) 
    changed_dictionary[col]=new_col
print(changed_dictionary)  
autos.rename(changed_dictionary,axis=1,inplace=True)
```

```python
brands=autos['brand'].value_counts().head(20).index
print(brands)
```

```python
brand_mean_price={}
for brand in brands:
    brand_prices=autos[autos['brand']==brand]
    mean_price=brand_prices['price'].mean()
    brand_mean_price[brand]=mean_price
print(brand_mean_price)
```

```python
brand_mean_mileage={}
for brand in brands:
    brand_mileage=autos[autos['brand']==brand]
    mean_mileage=brand_mileage['odometer_km'].mean()
    brand_mean_mileage[brand]=mean_mileage
print(brand_mean_mileage)
```

```python
mean_mileage=pd.Series(brand_mean_mileage)
mean_price=pd.Series(brand_mean_price)
```

```python
brand_analysis=pd.DataFrame(mean_price,columns=['mean_price'])
brand_analysis['mean_mileage']=mean_mileage
print(brand_analysis)
```
